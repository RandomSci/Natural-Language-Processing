```python
from gTTs import gTTS
import os
import pygame

s = 'hello, this is an example text for testing speech recognition'

def ok(text):
  tts = gTTS(text)
  tts.save("output.mp3")
  pygame.mixer.init()
  pygame.mixer.music.play()
  while pygame.mixer.music.get_busy():
    pygame.time.Clock().tick(10)

if __name__ == '__main__':
  s
```
